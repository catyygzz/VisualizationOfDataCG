```python
df = pd.read_csv("datasets/census.csv")
df.head(2)
```

```python
df.describe()
```

```python
x = df.columns.tolist()
x.sort()
x
```

```python
pd.pivot_table(df,
               index = ['STNAME','CTYNAME'],
               aggfunc = ['median'],
               values = ['BIRTHS2010','BIRTHS2011'],
               margins = True,
               columns = 'SUMLEV'
              )
```
