```python
counts = df['sexo'].value_counts()
counts = counts.to_frame()
type(counts)
```

```python
counts['%'] = (counts.iloc[:, :]/counts.loc[:,'sexo'].sum(axis=0))*100
counts
```

```python
counts.columns
```

```python
x= list(counts['sexo'])
y= list(counts['%'])

plt.subplot(1,2,1)
plt.rcParams['figure.figsize'] = (10,7)
plt.bar(x,y, width =0.8, color = 'darkcyan')


                   
```
