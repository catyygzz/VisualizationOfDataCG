```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("antropometria-dataset-2.csv")
df.head()
```

```python
pivot_q1 = pd.pivot_table(df,
               index = 'desc_ent',
               #aggfunc = ['median'], #Funciones de pivot
               values = 'peso', #valores que queremos solamente
               #margins = True, # muestra todas las columnas
              )
pivot_q1
```

```python
pivot_q2 = pd.pivot_table(df,
               index = 'desc_ent',
               aggfunc = 'median', 
               values = 'peso', 
               columns = 'sexo'
              )
pivot_q2
```

```python
pivot_q3 = pd.pivot_table(df,
               index = 'desc_ent',
               aggfunc = 'mean', 
               values = 'peso', 
               columns = 'sexo'
              )
pivot_q3
```

```python
df_adult = pd.read_csv("adult.csv")
df_adult.head()
```

```python
pivot_q4 = pd.pivot_table(df_adult,
                          index = "marital-status",
                          values= "age",
                          aggfunc= "mean",
                          columns= "education"
                         )
pivot_q4
```

```python
pivot_q5 = pd.pivot_table(df_adult,
                          index = "marital-status",
                          values= "age",
                          aggfunc= "median",
                          columns= ("education", "sex"),
                         )
pivot_q5
```
